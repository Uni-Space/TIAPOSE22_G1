## Correlation MAP

```r
library(ggstatsplot)

ggstatsplot::ggcorrmat(
  data = df,
  cor.vars = c(all:adult, maxtemp:maxwind), ## note how the variables are getting selected
  cor.vars.names = c(
    "all",
    "female",
    "male",
    "young",
    "adult",
    "maxtemp",
    "RH",
    "maxwind"
  ),
  colors = c("#999999", "white", "#0072B2"),
  title = "Relationship between the variables",
  ggcorrplot.args = list(outline.color = "black"),
  hc.order = TRUE
)
```

```r
ggstatsplot::ggcorrmat(
  data = df,             
  cor.vars = c(all, female, male, young, adult, maxtemp, RH, maxwind), ## 
  output = "correlations",
  corr.method = "robust",
  digits = 3
)
```

## Outliers

```r
# first method (percentage)
```

```r
lower_bound <- quantile(df$all, 0.025)
lower_bound
```

```r
upper_bound <- quantile(df$all, 0.975)
upper_bound
```

```r
# get indexes
outlier_ind <- which(df$all < lower_bound | df$all > upper_bound)
outlier_ind
```

```r
# get values
df[outlier_ind, "all"]
```

```r
# get mean
mean(df$all)
```

```r
# second method
```

```r
lower_bound <- median(df$all) - 3 * mad(df$all, constant=1)
lower_bound
```

```r
upper_bound <- median(df$all) + 3 * mad(df$all, constant=1)
upper_bound
```

```r
dates <- as.Date(df$date, format="%Y-%m-%d")
all <- df$all
plot(dates, all, type = "l", lty = 1)
```

```r
## Import library
library(ggplot2)
# Create plot
ggplot(data=df, aes(x=date, y=all, group=1)) +
  geom_line(color="red")+
  geom_point()
```
